```python
import numpy as np
import pandas as pd

from os.path import join
ruta = 'C:\Diplomado_IA_PUCP\Visualizacion_de_datos\proyecto'
```

```python
sinadef = pd.read_csv(join(ruta,'fallecidos_sinadef.csv'),sep=';',index_col=0,skiprows=2,encoding='latin1')[['SEXO','EDAD','PAIS DOMICILIO','DEPARTAMENTO DOMICILIO','PROVINCIA DOMICILIO','FECHA', 'AÑO', 'MES','MUERTE VIOLENTA']]
sinadef_covid = sinadef[(sinadef['AÑO']*100+sinadef['MES']>=202003) & (sinadef['MUERTE VIOLENTA']=='SIN REGISTRO') & (sinadef['FECHA']<='2021-05-20')]
```

```python
sinadef_covid['tmp'] = 1
```

```python
sinadef_covid_total = sinadef_covid[['FECHA','DEPARTAMENTO DOMICILIO','tmp']].groupby(['FECHA','DEPARTAMENTO DOMICILIO']).count().reset_index()
sinadef_covid_total
```

```python
##EXPORTANDO ,PARA UN RAPIDO USO EN EL DASHBOARD
sinadef_covid_total.to_csv(join(ruta,'sinadef_timeline.csv'),index=False)
```

```python
sinadef_covid_total
```

```python
sinadef_covid_total_filter = sinadef_covid_total = sinadef_covid_total[['FECHA','tmp']].groupby(['FECHA']).sum().reset_index().sort_values(by=['FECHA'])
sinadef_covid_total_filter
```

```python
import plotly.express as px

fig = px.line(sinadef_covid_total_filter, x="FECHA", y="tmp", title='Exceso de muertes')
fig.show()
```

```python
sinadef_covid_total_filter
```

```python
sinadef_covid_total_filter['avg_7'] = sinadef_covid_total_filter.tmp.rolling(7).mean()
```

```python
sinadef_covid_total_filter
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=sinadef_covid_total_filter['FECHA'], y=sinadef_covid_total_filter['tmp'],
                    mode='lines',
                    name='lines',
                    line = dict(color='red', width=0.25, dash='dash'),
                      hoverinfo='none'))
fig.add_trace(go.Scatter(x=sinadef_covid_total_filter['FECHA'], y=sinadef_covid_total_filter['avg_7'],
                    mode='lines',
                    name='lines',
                    line = dict(color='red', width=4)))

fig.update_layout(legend=dict(
    yanchor="top",
    y=1,
    xanchor="left",
    x=0
))
```
